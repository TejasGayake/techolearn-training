```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
students = np.array([[1,35],[2,40],[2,42],[3,50],[6,70],[7,75],[8,80],[9,85]])
```

```python
wcss = []

# 8-2centroids  = 6
for k in range(1,6):
    kmeans = KMeans(n_clusters = k,random_state=34)
    kmeans.fit(students)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),wcss,marker = "*")
plt.xlabel("Number of Clusters K")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```
