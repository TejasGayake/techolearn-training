```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
income = ['High','Low','Medium','Low','Medium','High']
job_type = ['Permanent','Temporary','Permanent','Temporary','Permanent','Permanent']
existing_login = ['No','Yes','No','Yes','Yes','No']
loan_status = ['Approved','Rejected','Approved','Rejected','Rejected','Approved']
```

```python
# Encode the values

le_income = LabelEncoder()
le_job_type = LabelEncoder()
le_loan = LabelEncoder()
le_loan_status = LabelEncoder()
```

```python
# we have to use fit_transform not only fit

income_encoded = le_income.fit_transform(income)
le_job_type_encoded = le_job_type.fit_transform(job_type)
le_loan_encoded = le_loan.fit_transform(existing_login)
le_loan_status_encoded = le_loan_status.fit_transform(loan_status)
```

```python
x = list(zip(income_encoded,le_job_type_encoded,le_loan_encoded))
y = le_job_type_encoded
```

```python
model = DecisionTreeClassifier()
model.fit(x,y)

new_income = le_income.transform(['High'])
new_job = le_job_type.transform(['Permanent'])
new_existing_loan = le_loan.transform(['No'])

new_customer = [[new_income[0],new_job[0],new_existing_loan[0]]]
```

```python
predictions = model.predict(new_customer)
final_result = le_loan_status.inverse_transform(predictions)
```
