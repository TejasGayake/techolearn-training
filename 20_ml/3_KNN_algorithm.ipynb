```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

x = np.array([
    [2,60],
    [4,65],
    [6,70],
    [8,80],
    [9,95]
    ])
```

```python
y = np.array([0,0,1,1,1])
```

```python
new_student = np.array([[11,79]])
```

```python
euclidean_model = KNeighborsClassifier(
    n_neighbors= 3,
    metric = 'euclidean'
) 

euclidean_model.fit(x,y)

euclidean_prediction = euclidean_model.predict(new_student)

print("Prediction : ",euclidean_prediction)
```

```python
# default model ---> euclidean


manhattan_model = KNeighborsClassifier(
    n_neighbors = 3,
    metric = "manhattan"
)
```

```python
manhattan_model.fit(x,y)

manhattan_model_prediction = manhattan_model.predict(new_student)

print("Prediction : ",manhattan_model_prediction)
```

```python
# default model ---> euclidean


minkowski_model = KNeighborsClassifier(
    n_neighbors = 3,
    metric = "minkowski"
)
```

```python
minkowski_model.fit(x,y)

minkowski_model_prediction = minkowski_model.predict(new_student)

print("Prediction : ",minkowski_model_prediction)
```
