```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = {
    "Study_Hours" : [1.5,2,2.5,3,3.5,4,4.5],
    "Mock_Tests" : [1,2,2,3,4,5,6],
    "Final_Marks" : [35,40,45,50,60,70,80]
}

df = pd.DataFrame(data)
```

```python
x = df[["Study_Hours", "Mock_Tests"]]
y = df["Final_Marks"]
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_:.4f}")
```

```python


pred = pd.DataFrame([[3.2, 3]], columns=["Study_Hours", "Mock_Tests"])
marks = model.predict(pred)
print(f"Predicted Marks: {marks[0]:.2f}")
```

```python

y_pred = model.predict(x_test)

plt.figure()
plt.scatter(y_test, y_pred, color="blue")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "r--")
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.title("Actual vs Predicted Values")
plt.tight_layout()
plt.show()
```
